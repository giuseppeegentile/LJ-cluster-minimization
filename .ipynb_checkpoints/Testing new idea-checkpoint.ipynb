```python
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math 

# Voxel grid parameters
grid_size = 10  # Define the size of the grid
voxel_size = 3.6  # Diameter of the larger atom (Kr)

# Initialize the voxel grid with random placements of atoms
def initialize_grid(grid_size, num_atoms):
    grid = np.zeros((grid_size, grid_size, grid_size), dtype=int)
    atom_positions = []
    atom_types = []  # Track the type of each atom
    for _ in range(num_atoms):
        x, y, z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
        while grid[x, y, z] != 0:
            x, y, z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
        grid[x, y, z] = 1  # 1 indicates an atom is present
        atom_positions.append((x, y, z))
        atom_types.append(random.choice(['Ar', 'Kr']))  # Randomly assign Ar or Kr to each position
    return grid, atom_positions, atom_types

# Calculate the distance between two atoms in voxel grid's unit of measurment
def distance(atom1, atom2):
    return np.sqrt(np.sum((np.array(atom1) - np.array(atom2))**2))

# Calculate the Lennard-Jones potential for a pair of atoms
def lennard_jones_potential(r, epsilon, sigma):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

# Define LJ parameters for Argon (Ar) and Krypton (Kr)
LJ_params = {
    'Ar': {'epsilon': 0.26, 'sigma': 3.82},
    'Kr': {'epsilon': 0.359, 'sigma': 4.09}
}

# Calculate the total Lennard-Jones potential for the system
def total_lennard_jones_potential(atom_positions, atom_types, LJ_params, voxel_size):
    total_potential = 0
    for i in range(len(atom_positions)):
        for j in range(i+1, len(atom_positions)):
            r = distance(atom_positions[i], atom_positions[j]) * voxel_size # distance in voxel grid multiplied by voxel size, Ã… unit of measure
                                                                            # to ensure consistent unit of measure for potential calculation
            if r > 1e-12:  # Avoid division by zero or extremely small distances
                epsilon_i = LJ_params[atom_types[i]]['epsilon']
                epsilon_j = LJ_params[atom_types[j]]['epsilon']
                if epsilon_i == epsilon_j:
                    sigma = LJ_params[atom_types[i]]['sigma']
                    epsilon = epsilon_i
                else:
                    epsilon = math.sqrt(epsilon_i*epsilon_j)
                    sigma = 0.5*(LJ_params['Ar']['sigma']+LJ_params['Kr']['sigma'])
                total_potential += lennard_jones_potential(r, epsilon, sigma)
    return total_potential
```

```python
mutation_probability = 0.15
selection_percentage = 3   # 1/3 are selected
```

```python
random.seed(119)
# Genetic Algorithm for optimization
def hybrid(grid_size, num_atoms, LJ_params, voxel_size, num_generations=500, population_size=20, folder_name = "GASA/iteration", sa_iterations=100,temperature=1.0, cooling_rate=0.99):
    # Initialize population
    population = [initialize_grid(grid_size, num_atoms) for _ in range(population_size)]
    history_evolution = []
    
    for generation in range(num_generations):

        # Calculate fitness
        fitness = [total_lennard_jones_potential(individual[1], individual[2], LJ_params, voxel_size) for individual in population]

        # Selection
        fitness_population = sorted(zip(fitness, population), key=lambda x: x[0])
        population = [x[1] for x in fitness_population[:population_size//selection_percentage]]
        if generation % 10 == 0:
            print("generation {}: fitness: {}".format(generation, min(fitness)))
            history_evolution.append(population[0])
            pop_npy = np.array(population[0])
            np.save(folder_name+ str(generation) + ".npy", pop_npy)
        
        # Crossover
        new_population = population[:]
        # crossover version 1
        if generation < 300:
            while len(new_population) < population_size:
                parent1, parent2 = random.sample(population, 2)
                crossover_point = random.randint(0, num_atoms-1)
                child_positions = parent1[1][:crossover_point] + parent2[1][crossover_point:]
                child_types = parent1[2][:crossover_point] + parent2[2][crossover_point:]
                new_population.append((None, child_positions, child_types))
        else: # crossover version 2
            alpha = []
            for p in range(len(population)):
                alpha.append(fitness[p])
            alpha = alpha / sum(alpha)
            
            while len(new_population) < population_size:
                parent1, parent2 = random.choices(population,
                                                  weights = alpha, 
                                                  k = 2)
                crossover_point = random.randint(0, num_atoms-1)
                child_positions = parent1[1][:crossover_point] + parent2[1][crossover_point:]
                child_types = parent1[2][:crossover_point] + parent2[2][crossover_point:]
                new_population.append((None, child_positions, child_types))
            
        
        # Mutation
        for individual in new_population:
            if random.random() < mutation_probability:  # Mutation probability
                idx = random.randint(0, num_atoms-1)
                individual[1][idx] = (random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1))
                individual[2][idx] = random.choice(['Ar', 'Kr'])  # Randomly assign new type
        
        population = [list(p) for p in new_population]


        # Apply SA to each individual periodically
        if generation % 100 == 0 and generation != 0:
            for i in range(len(population)):
                current_positions = population[i][1]
                current_potential = total_lennard_jones_potential(population[i][1], population[i][2], LJ_params, voxel_size)
                best_positions    = current_positions[:]
                best_potential    = current_potential
                new_temperature       = temperature
                history_lj = []
                grid = np.zeros((grid_size, grid_size, grid_size), dtype=int)

                # Place atoms at the specified positions
                for pos in population[i][1]:
                    x, y, z = pos
                    grid[x, y, z] = 1
                for iteration in range(sa_iterations):
                    new_positions = current_positions[:]
                    # Randomly select an atom and move it to a new position
                    atom_index = random.randint(0, len(new_positions) - 1)
                    new_x, new_y, new_z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
                    while grid[new_x, new_y, new_z] != 0:
                        new_x, new_y, new_z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
                    new_positions[atom_index] = (new_x, new_y, new_z)
                    
                    # Calculate the new potential
                    new_potential = total_lennard_jones_potential(new_positions, population[i][2], LJ_params, voxel_size)
                    
                    # Acceptance criterion
                    delta_potential = new_potential - current_potential
                    if delta_potential < 0 or random.uniform(0, 1) < math.exp(-delta_potential / new_temperature):
                        current_positions = new_positions
                        current_potential = new_potential
                        if new_potential < best_potential:
                            best_positions = new_positions
                            best_potential = new_potential
                      
                    # Decrease the temperature
                    new_temperature *= cooling_rate
                    if iteration % 10 == 0:
                        print(f"SA for individual {i} of generation {generation}: {iteration}: LJ potential {current_potential}, temperature: {new_temperature}")
                temperature *= cooling_rate
    
    
    # Get the best solution
    fitness = [total_lennard_jones_potential(individual[1], individual[2], LJ_params, voxel_size) for individual in population]
    best_solution = population[np.argmin(fitness)]

    return best_solution[1], best_solution[2], np.min(fitness), history_evolution

num_atoms = 30
best_positions, best_types, best_fitness, history_evolution = hybrid(grid_size, num_atoms, LJ_params, voxel_size)
print(f"Best solution: {best_positions}")
print(f"Atom types: {best_types}")
print(f"Best Lennard-Jones potential: {best_fitness}")
```

```python
# Genetic Algorithm with Simulated Annealing (GASA)
def GASA(grid_size, num_atoms, population_size, generations, T_init, T_final, alpha):
    # Initialize population
    population = [initialize_grid(grid_size, num_atoms) for _ in range(population_size)]
    best_solution = None
    best_fitness = float('inf')
    
    for gen in range(generations):
        # Calculate fitness for each individual
        fitness = [total_lennard_jones_potential(atom_positions, atom_types, LJ_params, voxel_size) 
                   for _, atom_positions, atom_types in population]
        
        # Selection (select the top 50% solutions)
        sorted_population = [x for _, x in sorted(zip(fitness, population))]
        selected_population = sorted_population[:population_size // 2]
        
        # Update best solution
        if min(fitness) < best_fitness:
            best_fitness = min(fitness)
            best_solution = selected_population[0]
        
        # Crossover and Mutation to create new population
        new_population = []
        while len(new_population) < population_size:
            # Select two parents
            parent1 = random.choice(selected_population)
            parent2 = random.choice(selected_population)
            
            # Crossover
            _, atom_positions1, atom_types1 = parent1
            _, atom_positions2, atom_types2 = parent2
            crossover_point = random.randint(1, num_atoms - 1)
            child_positions = atom_positions1[:crossover_point] + atom_positions2[crossover_point:]
            child_types = atom_types1[:crossover_point] + atom_types2[crossover_point:]
            
            # Mutation
            for _ in range(random.randint(1, 3)):  # Apply 1-3 random mutations
                mutate_index = random.randint(0, num_atoms - 1)
                x, y, z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
                while (x, y, z) in child_positions:
                    x, y, z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
                child_positions[mutate_index] = (x, y, z)
                child_types[mutate_index] = random.choice(['Ar', 'Kr'])
            
            new_population.append((np.zeros((grid_size, grid_size, grid_size), dtype=int), child_positions, child_types))
        
        # Apply Simulated Annealing
        T = T_init * (alpha ** gen)
        if T < T_final:
            break
        for i in range(population_size):
            _, atom_positions, atom_types = new_population[i]
            current_potential = total_lennard_jones_potential(atom_positions, atom_types, LJ_params, voxel_size)
            x, y, z = random.randint(0, grid_size-1), random.randint(0, grid_size-1), random.randint(0, grid_size-1)
            new_positions = atom_positions.copy()
            new_positions[random.randint(0, num_atoms - 1)] = (x, y, z)
            new_potential = total_lennard_jones_potential(new_positions, atom_types, LJ_params, voxel_size)
            delta_E = new_potential - current_potential
            if delta_E < 0 or random.random() < np.exp(-delta_E / T):
                new_population[i] = (np.zeros((grid_size, grid_size, grid_size), dtype=int), new_positions, atom_types)
        
        population = new_population
        if gen % 10 == 0:
            print(best_fitness)
    
    return best_solution, best_fitness
```

```python

# Parameters for GASA
population_size = 20
generations = 500
T_init = 1.0
T_final = 0.01
alpha = 0.99

# Run GASA
best_solution, best_fitness = GASA(grid_size, num_atoms, population_size, generations, T_init, T_final, alpha)
print("Best fitness:", best_fitness)
print("Best solution:", best_solution)

# Visualize the best solution
_, best_positions, best_types = best_solution
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for (x, y, z), atom_type in zip(best_positions, best_types):
    color = 'r' if atom_type == 'Ar' else 'b'
    ax.scatter(x, y, z, c=color, s=100)
plt.show()
```
